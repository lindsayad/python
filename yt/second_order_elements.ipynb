# Short Demonstration of Second Order Elements in yt

```python
import yt
```

```python
data_dir = "SecondOrderTets"
data_set = "few-element-mesh_out.e"
# data_set = "tet10_unstructured_out.e"
ds = yt.load(data_dir + "/" + data_set, step=-1)
```

## First a simple single element

```python
# create a default scene
sc = yt.create_scene(ds)
#sc = yt.create_scene(ds, field='uz')

# override the default colormap
ms = sc.get_source(0)
ms.cmap = 'Eos A'
ms.color_bounds = (.07, .15)
# ms.color_bounds = (-.02, .02)
# ms.color_bounds = (-.01, .2)

# adjust the camera position and orientation
cam = sc.camera
# increase the default resolution
cam.resolution = (800, 800)

cam.focus = ds.arr([0.5, 0.5, 0.5], 'code_length')

cam_pos = ds.arr([0.5, 0.5, -1.0], 'code_length')
north_vector = [1,0,0]
cam.set_position(cam_pos, north_vector)
```

### Viewing tet from plus z-axis. You can see the curved contours of the field denoting the second order interpolation

```python
sc.show()
```

## Now we'll briefly examine an actual data-set.

### This data set contains a Navier stokes simulation with flow through a pipe that has an exit diameter larger than its input diameter

```python
data_dir = "SecondOrderTets"
data_set = "constant_ending_channel_3d_unstructured_out.e"
ds = yt.load(data_dir + "/" + data_set, step=-1)
```

```python
# create a default scene
sc = yt.create_scene(ds, field='uz')

# override the default colormap
ms = sc.get_source(0)
ms.cmap = 'Eos A'
ms.color_bounds = (-.05, .25)

# adjust the camera position and orientation
cam = sc.camera
# increase the default resolution
cam.resolution = (800, 800)

cam.focus = ds.arr([0.5, 0.5, 0.5], 'code_length')

cam_pos = ds.arr([1, 1, 5.0], 'code_length')
#north_vector = [1,0,0]
cam.set_position(cam_pos, north_vector)
```

### Looking at the outlet profile (slightly tiliting the geometry). The element curvature is nicely captured

```python
sc.show()
```

```python
# create a default scene
sc = yt.create_scene(ds, field='uz')

# override the default colormap
ms = sc.get_source(0)
ms.cmap = 'Eos A'
ms.color_bounds = (-.1, 1)

# adjust the camera position and orientation
cam = sc.camera
# increase the default resolution
cam.resolution = (800, 800)

cam.focus = ds.arr([0.5, 0.5, 0], 'code_length')

cam_pos = ds.arr([0.5, 0.5, -1], 'code_length')
north_vector = [1,0,0]
cam.set_position(cam_pos, north_vector)
```

### Now the inlet profile

```python
sc.show()
```

```python
# create a default scene
sc = yt.create_scene(ds, field='uz')

# override the default colormap
ms = sc.get_source(0)
ms.cmap = 'Eos A'
ms.color_bounds = (-.1, .1)

# adjust the camera position and orientation
cam = sc.camera
# increase the default resolution
cam.resolution = (800, 800)

cam.focus = ds.arr([0.5, 0.5, 2], 'code_length')

cam_pos = ds.arr([1.5, 0.5, 2], 'code_length')
north_vector = [0,1,0]
cam.set_position(cam_pos, north_vector)
```
